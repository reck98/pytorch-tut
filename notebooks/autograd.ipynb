```python
import math

# input
x = 2.0

# true label (binary)
y = 1

# parameters
w = 0.5
b = 0.1

# --------------------
# Forward Pass
# --------------------

# linear output
z = w * x + b

# sigmoid activation
y_pred = 1 / (1 + math.exp(-z))

# binary cross entropy loss
loss = -(y * math.log(y_pred) +
         (1 - y) * math.log(1 - y_pred))

print("Prediction:", y_pred)
print("Loss:", loss)

# --------------------
# Backpropagation
# --------------------

# derivative of loss wrt z
dL_dz = y_pred - y

# gradients
dL_dw = dL_dz * x
dL_db = dL_dz

print("Gradient wrt w:", dL_dw)
print("Gradient wrt b:", dL_db)

# --------------------
# Gradient Descent
# --------------------

learning_rate = 0.1

w = w - learning_rate * dL_dw
b = b - learning_rate * dL_db

print("Updated weight:", w)
print("Updated bias:", b)
```

```python
# input
x = 2.0

# true label (binary)
y = 1

# parameters
w = 0.5
b = 0.1

x = torch.tensor(x)
y = torch.tensor(y)
w = torch.tensor(w, requires_grad=True)
b = torch.tensor(b, requires_grad=True)
```

```python
def binary_cross_entropy_loss(y_pred, y):
    return -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred))
```

```python
## simple nn

z = w * x + b
y_pred = torch.sigmoid(z)

loss = binary_cross_entropy_loss(y_pred, y)

loss.backward()

print("Prediction:", y_pred)
print("Loss:", loss)
print("Gradient wrt w:", w.grad)
print("Gradient wrt b:", b.grad)

learning_rate = 0.1

w = w - learning_rate * w.grad
b = b - learning_rate * b.grad

print(f"Updated weight: {w}")
print(f"Updated bias: {b}")
```
