```python
import torch
import torch.nn as nn
```

```python
class MyModel(nn.Module):
    
    def __init__(self, num_features):
        
        super().__init__()
        
        self.linear = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        
        self.hidden = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        
        x = self.linear(x)
        x = self.relu(x)
        
        x = self.hidden(x)
        x = self.sigmoid(x)
        
        return x

    
        
```

```python
## make a dataset 

features = torch.rand((100, 5))
labels = torch.rand((100, 1))
print(features.shape)
print(labels.shape)
```
